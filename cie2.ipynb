```python
import numpy as np
from scipy import stats
data = pd.read_csv("/content/titanic.csv")
print(data)
mean1=np.mean(data['Age'])
medain1=np.median(data['Age'])
mode1=stats.mode(data['Age'],keepdims=True).mode[0]
print('mean of age is',mean1)
print('median of age is',medain1)
print('mode of age is',mode1)
mean2=np.mean(data['Fare'])
medain2=np.median(data['Fare'])
mode2=stats.mode(data['Fare'],keepdims=True).mode[0]
print('mean of fare is',mean2)
print('median of fare is',medain2)
print('mode of fare is',mode2)
```

```python
import pandas as pd
data['Servived']=data['Survived'].replace({1:'servived',0:'not servived'})
print(data['Servived'])

import matplotlib.pyplot as plt
grouped_data=data.groupby(['Servived','Pclass'])['PassengerId'].count().unstack()
grouped_data.plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(ticks=[0,1],labels=[0,1],rotation=0)
plt.title('Passenger Count by Survived status')
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy import stats
mean=np.mean(data['Age'])
std=np.std(data['Age'])
print('mean of age is',mean)
print('std. deviation is',std)

```

```python
threshold=3
outlier=[]
for i in data['Age']:
  z=(i-mean)/std
  if z>   threshold:
    outlier.append(i)
  print(outlier)
```
